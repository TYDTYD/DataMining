```python
import sys
from collections import defaultdict
from itertools import tee

class SegmentTree:
    def __init__(self, n):
        self.n = n
        self.seg = [-float('inf')] * (2 * n)
    
    def update(self, i, val):
        i += self.n
        self.seg[i] = val
        while i > 1:
            i //= 2
            self.seg[i] = max(self.seg[2 * i], self.seg[2 * i + 1])
    
    def query(self, l, r):
        mx = -float('inf')
        l += self.n
        r += self.n + 1
        while l < r:
            if l & 1:
                mx = max(mx, self.seg[l])
                l += 1
            if r & 1:
                r -= 1
                mx = max(mx, self.seg[r])
            l //= 2
            r //= 2
        return mx

def dist(V, N):
    # Preprocessing coordinates compression
    tmp = [y for _, y, _ in V]
    tmp = sorted(set(tmp))
    compCoor = {v: i for i, v in enumerate(tmp)}
    
    # Sort input by x, then by y
    V.sort()

    # Initialize distances
    distances = [float('inf')] * N
    seg_tree = SegmentTree(len(compCoor))

    # Set distances for overlapping points to 0
    for i in range(N - 1):
        if V[i][0] == V[i + 1][0] and V[i][1] == V[i + 1][1]:
            distances[V[i][2]] = 0
            distances[V[i + 1][2]] = 0
    
    # Segment tree process
    for x, y, idx in V:
        y_trans = compCoor[y]
        val = seg_tree.query(0, y_trans)
        distances[idx] = min(distances[idx], x + y - val)
        seg_tree.update(y_trans, x + y)

    return distances

def solve(N, V):
    # Initialize distances to a very large value
    distances = [float('inf')] * N
    
    # Perform the algorithm four times to consider different quadrants
    for _ in range(4):
        # Calculate distances in this configuration
        dist_vals = dist(V, N)
        # Merge calculated distances
        distances = [min(distances[i], dist_vals[i]) for i in range(N)]
        
        # First flip x to -x
        for i in range(N):
            V[i] = (-V[i][0], V[i][1], V[i][2])
        
        # Second flip y to -y after the first run
        if _ == 1:
            for i in range(N):
                V[i] = (V[i][0], -V[i][1], V[i][2])
    
    return distances


N = len(df)
V = []
i=0

for _, row in df.iterrows():
    V.append((row['위도'], row['경도'],i))
    i+=1
    
result = solve(N, V)
```

```python
# 여전히 서울 밖 데이터가 존재
def remove_outlier_radius(df, column1, column2, radius):
    seoul_lat = 37.5642135
    seoul_lon = 127.0016985
    
    distance = ((df[column1] - seoul_lat) ** 2 + (df[column2] - seoul_lon) ** 2) ** 0.5
    
    filtered_df = df[distance < radius]
    return filtered_df

df = remove_outlier_radius(df, '위도', '경도', 0.2)
```
